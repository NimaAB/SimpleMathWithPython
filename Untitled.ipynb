```python
#importing python libreries.

import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
#plot of numbers 1:
for j in range(0,10):
    x = j**2
    plt.plot(j,x,"o")
```

```python
#plot with numPy libreries: 

x_1 = np.arange(0,10,0.001)
y = np.sqrt(x_1)
plt.plot(x_1,y,"-")

func = (*map(lambda x: x**2,x_1),) 
plt.plot(x_1,func,"-")

```
